```python
video=cv2.VideoCapture('test1.mp4')
writer = cv2.VideoWriter('output.mp4')
net, output_layers, classes = loading_dependencies()
```

```python
def calculateCentroid(xmin,ymin,xmax,ymax):

    xmid = ((xmax+xmin)/2)
    ymid = ((ymax+ymin)/2)
    centroid = (xmid,ymid)

    return xmid,ymid,centroid
```

```python
def draw_detection_box(frame,x1,y1,x2,y2,color):

    cv2.rectangle(frame,(x1,y1),(x2,y2), color, 2)
```
